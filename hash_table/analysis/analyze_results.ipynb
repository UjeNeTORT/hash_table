```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
def draw_load_hist (fpath):
    df = pandas.read_csv(fpath)
    nparr = df.to_numpy ()
    arr1 = [i[1] for i in nparr]
    arr0 = [i[0] for i in nparr]

    plt.bar (arr0, arr1, width=5)
    plt.show ()
```

```python
draw_load_hist ("./results/wd_len_hf.csv")
```
