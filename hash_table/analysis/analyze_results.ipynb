```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
def draw_load_hist (fpath, hf_name):
    df       = pandas.read_csv(fpath)
    nparr    = df.to_numpy ()
    ids      = [i[0] for i in nparr]
    l_length = [i[1] for i in nparr]

    plt.title ("load historgram for " + hf_name + " hash function")
    plt.xlabel ("id")
    plt.ylabel ("corresponding list length")
    plt.grid (linestyle= ":")
    plt.bar (ids, l_length, width=5)
    plt.show ()

    return np.sqrt (np.std (l_length)), np.mean (l_length)
```

## Построение гистограмм

```python
draw_load_hist ("./results/csv/retzero_hf.csv", "\"always return 0\"")
```

```python
draw_load_hist ("./results/csv/ascii_first_hf.csv", "\"ascii code of first letter\"")
```

```python
draw_load_hist ("./results/csv/wd_len_hf.csv", "\"length of the word\"")
```

```python
draw_load_hist ("./results/csv/ctrl_sum_hf.csv", "control sum")
```

```python
draw_load_hist ("./results/csv/sum_by_len_hf.csv", "\"control sum divided by length\"")
```

```python
draw_load_hist ("./results/csv/crc32_hf.csv", "crc32")
```
